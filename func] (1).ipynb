```python
def cacl():
    print("Calculator")
    print(f"+ for addition")
    print(f"- for subtraction")
    print(f"* for multiply")
    num1=int(input("Enter first number"))
    num2=int(input("Enter second number"))
    choice=input("enter your op +,-,*")
    if choice=='+':
      add(num1,num2)
    elif choice=='-':
      subt(num1,num2)
    elif choice=='*':
      mul(num1,num2)
    else:
     print("Enter a valid op")
```

```python
passwords=[]
users=[]
```

```python
username=input("Enter username")
password=input("Enter password")
if len(password)<6:
    print("password is too short")
elif ' ' in password:
   print("no space allowed")
elif not any (i in '!@#$%^&*()_+' for i in password):
     print("password must have special character")
elif not any(char.isupper() for char in password):
    print("One UpperCase is required")
elif not any(char.islower() for char in password):
     print("Onr lowercase is required")
elif not any(char.isdigit() for char in password):
     print("One digit is required")
else:
    print("Pasword is strong")
    passwords.append(password)
    users.append(username)
```
