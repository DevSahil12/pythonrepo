```python
class bankaccount:
    def __init__(self,name,password,balance=0):
        self.name=name
        self.password=password
        self.balance=balance
        self.transactions=[]
        self.transactions.append(f"name:-{name}")
        self.transactions.append(f"opening balance={balance}")
        print("welcome to our bank")
    def check_password(self, password):
        if len(password) < 6:
            print("Password must be at least 6 characters")
            return False
        elif not re.search(r"[A-Z]", password):
            print("At least one uppercase letter required")
            return False
        elif not re.search(r"[a-z]", password):
            print("At least one lowercase letter required")
            return False
        elif not re.search(r"\d", password):
            print("At least one digit required")
            return False
        elif not re.search(r"[!@#$%^&*]", password):
            print("At least one special character required")
            return False
        else:
            print("password is strong")
            return True
    def auth(self):
         passw=input("Enter your password")
         if passw==self.password:
             print("Login successfully")
             return True
         else:
             print("Wrong password")
             return False
    def deposit(self,amount):
        if self.auth():
          if amount>0:
              self.balance+=amount
              self.transactions.append(f"deposited amount is {amount} and balance is {self.balance}")
          else:
              print("amount needs to be positive")
    def withdrawal(self,amount):
        if self.auth():
           if amount<self.balance:
               self.balance-=amount
               self.transactions.append(f"withdrawal amount:-{amount} and balance is {self.balance}")
           else:
               print("insufficient balance")
    def viewBalance(self):
        if self.auth():
            print(f"{self.name} , {self.balance}")
    def transaction(self):
        if self.auth():
            for i in self.transactions:
               print(i)
```

```python
class bankaccount1:
    def _init_(self):
        self.name=input('enter name for customer')
        self.balance=int(input('enter opening balance for account'))
        self.username=input('enter username')
        while True:
            self.password = input('Enter password: ')
            pattern = r'^(?=.[A-Z])(?=.[a-z])(?=.\d)(?=.[@$!%?&])[A-Za-z\d@$!%?&]{8,}$'
            if re.match(pattern, self.password):
                print("Strong password accepted")
                break
            else:
                print("Weak password!")
                print("Password must contain:")
                print("- Minimum 8 characters")
                print("- 1 uppercase letter")
                print("- 1 lowercase letter")
                print("- 1 digit")
                print("- 1 special character (@$!%*?&)")
        self.transactions=[]
        self.transactions.append(f"Name:- {self.name}")
        self.transactions.append(f"opening balance= {self.balance}")
        print('welcome to our bank')
    def deposit(self):
        self.user=input('enter username')
        self.passw=input('enter password')
        if self.user==self.username and self.passw==self.password:
            self.amount=int(input('enter amount to deposit'))
            if self.amount >0:
                self.balance+=self.amount
                self.transactions.append(f"deposit amount is {self.amount} and balance is {self.balance}")
            else:
                print('amount must be in positive')
        else:
            print('wrong username')
    def withdrawl(self, amount):
        self.user=input('enter username')
        self.passw=input('enter password')
        if self.user==self.username and self.passw==self.password:
            if amount<self.balance:
                self.balance-=amount
            else:
                print('insufficient balance')
            self.transactions.append(f"withdraw amount:- {amount} balance= {self.balance}")
        else:
            print('authentication failed')
    def viewbalance(self):
        print(f"{self.name}, {self.balance}")
    def transaction(self):
        for i in self.transactions:
            print(i)
```
