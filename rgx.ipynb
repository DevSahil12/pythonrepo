```python
def withdraw():
    amount=10000
    withdraw=int(input("Enter amount to withdraw"))
    if(withdraw<amount):
      amount-=withdraw
      print(f"Withdrawal is successfull rest amount {amount}")
    else:
       print("Enter lower amount deposited money is less")
```

```python
withdraw()
```
