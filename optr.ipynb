```python
choice = int(input())
match choice:
    case 1:
        print('one')
    case 2:
        print('two')
    case _:
        print('invalid')
```

```python
a=int(input('enter a'))
b=int(input('enter b'))
op=input('operation')
match op:
    case '+':
        print(a+b)
    case '-':
        print(a-b)
    case '*':
        print(a*b)
    case _:
        print('invalid')
```

```python
day=input("enter day")
match day:
    case 'sat' | 'sun':
        print("Holidays")
    case 'mon'|'tue'|'wed'|'thrus':
        print("working")
    case _:
        print("Not valid")
   
              
```
