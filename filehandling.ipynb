```python
user=['king','ring','sing']
password=['123','456','789']
```

```python
with open('auth.txt','w') as f:
    for u,p in zip(user,password):
        f.write(f"{u},{p}\n")
```

```python
user1=[]
password1=[]
```

```python
with open('auth.txt','r') as f:
    for line in f:
        u,p=line.strip().split(",")
        user1.append(u)
        password1.append(p)
```

```python
print(user1)
```

```python
print(password1)
```
